```python
import pyalps
import numpy as np
import matplotlib.pyplot as plt
import pyalps.plot
```

```python
parms = [ { 
           'LATTICE'                   : "open square lattice", 
           'MODEL'                     : "fermion Hubbard",
           'CONSERVED_QUANTUMNUMBERS'  : 'Nup, Ndown',
           'Nup_total'                  : 1,
           'Ndown_total'                  : 1,
           't'                         : 1,
           'U'                         : -7,
           'SWEEPS'                    : 4,
           'NUMBER_EIGENVALUES'        : 2,
           'L'                         : 80,
           'W'                         : 1,
           'MAXSTATES'                 : 200
          } ]
```

```python

a = list()
b = list()
c = list()
e1 = list()
e2 = list()
e3 = list()

#e2

parms[0]['Nup_total'] = 2
parms[0]['Ndown_total'] = 0

for k in range(30):
    input_file = pyalps.writeInputFiles('parm_spin_one_half',parms)
    res = pyalps.runApplication('dmrg',input_file,writexml=True)

    data = pyalps.loadEigenstateMeasurements(pyalps.getResultFiles(prefix='parm_spin_one_half'))
    for s in data[0]:
       b.append(s.y[0])
    
    parms[0]['Nup_total'] = parms[0]['Nup_total'] + 2
    parms[0]['Ndown_total'] = parms[0]['Ndown_total'] + 2
    
for m in range(0, len(b), 2):
    e2.append(b[m])

#e3
parms[0]['Nup_total'] = 2
parms[0]['Ndown_total'] = 2

for k in range(30):
    input_file = pyalps.writeInputFiles('parm_spin_one_half',parms)
    res = pyalps.runApplication('dmrg',input_file,writexml=True)

    data = pyalps.loadEigenstateMeasurements(pyalps.getResultFiles(prefix='parm_spin_one_half'))
    for s in data[0]:
       c.append(s.y[0])
    
    parms[0]['Nup_total'] = parms[0]['Nup_total'] + 2
    parms[0]['Ndown_total'] = parms[0]['Ndown_total'] + 2
    
for f in range(0, len(c), 2):
    e3.append(c[f])

    
#e1
parms[0]['Nup_total'] = 1
parms[0]['Ndown_total'] = 1
    
for x in range(30):
    input_file = pyalps.writeInputFiles('parm_spin_one_half',parms)
    res = pyalps.runApplication('dmrg',input_file,writexml=True)

    data = pyalps.loadEigenstateMeasurements(pyalps.getResultFiles(prefix='parm_spin_one_half'))
    for s in data[0]:
       a.append(s.y[0])
    
    parms[0]['Nup_total'] = parms[0]['Nup_total'] + 2
    parms[0]['Ndown_total'] = parms[0]['Ndown_total'] + 2

    

for n in range(0, len(a), 2):
    e1.append(a[n])
```

```python
h = list()
for j in range(len(e1)):
    h.append((e2[j] - e1[j])/2)
print(h)

mu = list()
for i in range(len(e1)):
    mu.append((e3[i] - e1[i])/2)
    
print(mu)
```
